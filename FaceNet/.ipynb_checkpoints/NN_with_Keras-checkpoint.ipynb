```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
# load the dataset
dataset = loadtxt('pima-indians-diabetes.csv', delimiter=',')
```

```python
#split dataset into input(X) and output(y)
X = dataset[:, 0:8]
y = dataset[:, 8]
```

```python
# Define the model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on dataset
model.fit(X, y, epochs=250, batch_size=15)
```

```python
from matplotlib import pyplot
def rectified(x):
    return max(0.0, x)
```

```python
series_in = [x for x in range(-5, 12)]
series_out = [ rectified(x) for x in series_in]
pyplot.plot(series_in, series_out)
pyplot.show()
```

```python
import os
from keras.models import model_from_json
```

```python
#serialize model to JSON
model_json = model.to_json()
with open("my_model.json", "w") as json_file:
    json_file.write(model_json)
# Serialize model to HDF5
model.save_weights("my_model.h5")
print("Saved model to disk")
```

```python
from keras.models import load_model
```

```python
# load json and create model
json_file = open('my_model.json', 'r')
loaded_model_from_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_from_json)
loaded_model.load_weights('my_model.h5')
print('Loaded model from disk')
```

```python
loaded_model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
score = loaded_model.evaluate(X, y, verbose=0)
print("%s: %.2f%%" % (loaded_model.metrics_names[1], score[1]*100))
```
